```r
library(tidyverse)
library(ggplot2)
library(MatchIt)
library(dplyr)
library(tidyr)
library(sandwich)
```

```r
# Loading Data
data <- read_csv("../datahwk2.1/final-data.csv")
data2018 <- read_csv("../datahwk2.1/data-2018.csv")
```

```r
# clean data
data <- data %>%
  select(-starts_with("year")) %>%
  mutate(year = data$year.x)
```

# Question 1: filter data and create a plan number box and whisker plot

```r
plan_type <- data %>%
  group_by(plan_type) %>%
  summarise(n_plans = n_distinct(planid)) %>%
  arrange(desc(n_plans))
```

```r
# Filter plans
filtered_data <- data %>%
  filter(
    snp != "Yes",
    !grepl("^8", as.character(planid)), 
    plan_type != "PDP",
      !is.na(planid)
  )

# Plan counts per county per year
county_plan_counts <- filtered_data %>%
  distinct(contractid, fips, year) %>%
  count(fips, year, name = "n_plans")

# Boxplot
ggplot(county_plan_counts, aes(x = factor(year), y = n_plans)) +
  geom_boxplot() + 
  labs(
    title = "Boxplot of distributon of plan counts by year (2014-2019)",
    x = "Year",
    y = "Number of MA Plans)") +
  theme_minimal()
```

A majority of counties have a sufficient number of plans, on average 5 to 10, some counties have far too many, as shown by the significant outlier in the dataset. 

# Question 2: Frequency histograms showing the dsitribution of plan bids

```r
# remove rows with NA
bids <- filtered_data %>%
    filter(!is.na(bid))

# Histogram for 2014
ggplot(filter(bids, year == 2014), aes(x = bid)) +
  geom_histogram(bins = 50) +
  labs(
    title = "Plan Bids (2014)",
    x = "Plan Bid",
    y = "Count") +
  theme_minimal()

# Histogram for 2018
ggplot(filter(bids, year == 2018), aes(x = bid)) +
  geom_histogram(bins = 50) +
  labs(
    title = "Plan Bids (2018)",
    x = "Plan Bid",
    y = "Count") +
  theme_minimal()
```

The distributions for both bids appear to be approximately normal, but the median for 2018 is significantly higher than 2014, and there is a slightly narrower spread for 2018. The peak for 2014 is slightly above 6000 in 2014 for aplan bid of slightly less than 800, while for 2018 is up above 8750 for a plan bid of slightly more than 700. As the number of competing plans grew firms maintained more competitive pricing.

# Question 3: Average HHI from 2014 to 2019

```r
# Compute HHI
county_hhi <- filtered_data %>%
  group_by(year, fips) %>%
  mutate(
    share = avg_enrollment / sum(avg_enrollment, na.rm = TRUE)
  ) %>%
  summarize(
    hhi = sum(share^2, na.rm = TRUE),
    .groups = "drop"
  )

# Average HHI by year
avg_hhi_year <- county_hhi %>%
  group_by(year) %>%
  summarize(
    avg_hhi = mean(hhi, na.rm = TRUE)
  )

# Plot HHI average
ggplot(avg_hhi_year, aes(x = year, y = avg_hhi)) +
  geom_line() +
  geom_point() +
  labs(
    title = "Average HHI from 2014 to 2019",
    x = "Year",
    y = "Average County HHI") +
  theme_minimal()
```

The avergae HHI has decreased steadily from 2014 to 2019. In 2014 average HHI started with an index of about 0.36 and fell to 0.3 by the 2019, although there was a slight peak in 2017 followd by increasing declines in 2018 and 2019. 

# Question 4: Average share of Medicare Advantage from 2014 to 2019

```r
county_penetration <- filtered_data %>%
  group_by(year, fips) %>%
  summarize(
    ma_share = mean(avg_enrolled / avg_eligibles, na.rm = TRUE),
    .groups = "drop"
  )

# Average by year
avg_penetration <- county_penetration %>%
  group_by(year) %>%
  summarize(
    avg_ma_share = mean(ma_share, na.rm = TRUE)
  )

# Plot of medicare advantage over time
ggplot(avg_penetration, aes(x = year, y = avg_ma_share)) +
  geom_line() +
  geom_point() +
  labs(
    x = "Year",
    y = "Average MA Share") +
  theme_minimal()
```

Medicare Advantage shares have increased steadily over time. Shares began at slightly above 0.21 in 2014 and reached just under 0.28 by 2019. The trend was continuously upwards during this period.

### load 2018 data

```r
# clean data
d18 <- data2018 %>%
  filter(
    snp != "Yes",        
    !grepl("^8", planid),
    plan_type != "PDP"
  )

market_shares <- d18 %>%
  group_by(fips, parent_org) %>%
  summarize(firm_enrollment = sum(avg_enrollment, na.rm = TRUE), .groups = "drop") %>%

# Calculate total market shares and sizes
group_by(fips) %>%
mutate(
    total_market_enrollment = sum(firm_enrollment),
    market_share = (firm_enrollment / total_market_enrollment) * 100
  ) %>%

# Calculate HHI for each fips
summarise(hhi = sum(market_share^2), .groups = "drop")

# Determine cutoffs
hhi_cutoffs <-quantile(market_shares$hhi, probs = c(0.33, 0.66), na.rm = TRUE)
p33 <- hhi_cutoffs["33%"]
p66 <- hhi_cutoffs["66%"]

# Merge HHI 
final2018 <- d18 %>%
 left_join(market_shares, by = "fips")%>%
mutate(
    market_type = case_when(
      hhi <= p33 ~ "Competitive",
      hhi >= p66 ~ "Uncompetitive",
      TRUE ~ "Moderate" 
    )
)
```

# Question 5: average bid among competitve and uncompetitve markets

```r
bid_comparison_table <- final2018 %>%
  filter(market_type %in% c("Competitive", "Uncompetitive")) %>%  
  group_by(market_type) %>%
  summarise(
    avg_bid = mean(bid, na.rm = TRUE),
    n_obs = n(),
    .groups = "drop"
  ) %>%
  
  mutate(
    avg_bid = round(avg_bid, 2)
  ) %>%
  
  rename(Market_Type = market_type)

bid_comparison_table
```

# Question 6: Split market into quartiles based on Medicare FFS costs

```r
# Quartile variables 
ffs_cuts <- quantile(unique(final2018$avg_ffscost), probs = c(0.25, 0.50, 0.75), na.rm = TRUE)
ffs_summary_table <- final2018 %>%
  filter(market_type %in% c("Competitive", "Uncompetitive")) %>%

  mutate(
    ffs_quartile_cat = case_when(
      avg_ffscost <= ffs_cuts[1] ~ 1,
      avg_ffscost <= ffs_cuts[2] ~ 2,
      avg_ffscost <= ffs_cuts[3] ~ 3,
      TRUE ~ 4
    )
  ) %>%
  
# Group Average Bid
  group_by(ffs_quartile_cat, market_type) %>%
  summarise(
    avg_bid = mean(bid, na.rm = TRUE), 
    .groups = "drop"
  ) %>%
  
  mutate(avg_bid = round(avg_bid, 2)) %>%
    pivot_wider(
    names_from = market_type, 
    values_from = avg_bid
  ) %>%
    mutate(
    Quartile_Label = case_when(
      ffs_quartile_cat == 1 ~ "Q1",
      ffs_quartile_cat == 2 ~ "Q2",
      ffs_quartile_cat == 3 ~ "Q3",
      ffs_quartile_cat == 4 ~ "Q4"
    )
  ) %>%
  select(Quartile_Label, Competitive, Uncompetitive)

ffs_summary_table
```

# Qustion 7: ATE with estimators

```r
# analysis dataset
analysis_df <- final2018 %>%
  filter(market_type %in% c("Competitive", "Uncompetitive"), !is.na(avg_ffscost), !is.na(bid)) %>%
  mutate(
    treat = as.numeric(market_type == "Uncompetitive"),
    ffs_quartile = factor(ntile(avg_ffscost, 4))
  )


get_stats <- function(m) {
  if (inherits(m, "lm")) {
    return(c(coef(m)["treat"], sqrt(vcovHC(m, type = "HC1")["treat", "treat"])))
  }
  d <- if(inherits(m, "matchit")) match.data(m) else m
  ate <- mean(d$bid[d$treat == 1]) - mean(d$bid[d$treat == 0])
  se <- sqrt(var(d$bid[d$treat == 1])/sum(d$treat == 1) + var(d$bid[d$treat == 0])/sum(d$treat == 0))
  return(c(ate, se))
}

# Matching
m_invvar <- matchit(treat ~ ffs_quartile, data = analysis_df, method = "nearest", distance = "glm")
m_mahal  <- matchit(treat ~ ffs_quartile, data = analysis_df, method = "nearest", distance = "mahalanobis")

# IPW and OLS
ps_fit <- glm(treat ~ ffs_quartile, data = analysis_df, family = binomial)
analysis_df <- analysis_df %>% 
  mutate(ps = predict(ps_fit, type = "response"),
         weight = ifelse(treat == 1, 1/ps, 1/(1-ps)))

ipw_mod <- lm(bid ~ treat, data = analysis_df, weights = weight)
ols_mod <- lm(bid ~ treat * ffs_quartile, data = analysis_df)

# Final Table
results_raw <- rbind(get_stats(m_invvar), get_stats(m_mahal), get_stats(ipw_mod), get_stats(ols_mod))
ate_results <- data.frame(
  Estimator = c("INV", "MAH", "IPW", "OLS"),
  ATE = round(results_raw[,1], 2),
  SE  = round(results_raw[,2], 2)
)

ate_results
```

# Question 8:
The inverse variance estimate is very low, while the Mahalanobis estimate is very high, the inverse propensity weighting and simple linear regression provide more moderate estimate in between the two. Overall, these estiamtes are fairly high other than the inverse variance estimate. Despite the differences in the avergae treatment effect estimators the standard errors are all similar with the linear regression standard error the highest at 2.43.

# Question 9: re-estimate treatment effects using the continuous FFS costs variable as well at total Medicare beneficiaries as covariates

```r
# 1. Re-estimate using continuous FFS and Total Beneficiaries (avg_eligibles)
ols_continuous <- lm(
  bid ~ treat + avg_ffscost + avg_eligibles, 
  data = analysis_df
)

# 2. Extract results for comparison
# Using your existing ols_mod (the one with interaction and quartiles)
res_quartile <- c(
  ATE = coef(ols_mod)["treat"], 
  SE = sqrt(vcovHC(ols_mod, type = "HC1")["treat", "treat"])
)

res_continuous <- c(
  ATE = coef(ols_continuous)["treat"], 
  SE = sqrt(vcovHC(ols_continuous, type = "HC1")["treat", "treat"])
)

# 3. Build Comparison Table
comparison_tab <- data.frame(
  Specification = c("OLS Baseline", "OLS Alternate"),
  ATE = round(c(res_quartile[1], res_continuous[1]), 2),
  SE = round(c(res_quartile[2], res_continuous[2]), 2)
)

comparison_tab
```

The OLS estimator with Medicare beneficiaries as covariates provides a more presise estimate, with a much lower standard error than the baseline estimate. Additionally, the average treatment effect estimate rose from 9.79 with the baseline estimator to 14.70 with the alternate estiamtor
