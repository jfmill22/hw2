```r
library(tidyverse)
library(ggplot2)
library(MatchIt)
library(fixest)
years <- 2014:2019
```

```r
# Loading Data
data <- read_csv("../data/final-data.csv")
data2018 <- read_csv("../data/data-2018.csv")
```

```r
glimpse(data)
```

```r
# 1. Filter plants: remove SNPs, 800-series, and prescription drug only
filtered_plans <- data %>%
  filter(
    snp != "Yes",        
    !grepl("^8", planid),
    plan_type != "PDP"
  )

filtered_plan_type <- filtered_plans %>%
  group_by(plan_type) %>%
  summarise(n_plans = n_distinct(planid)) %>%
  arrange(desc(n_plans))

filtered_plan_type
```

```r
# Box and whisker plot of plan counts by count over time
county_plan_counts <- data %>%
  distinct(contractid, fips, year) %>%
  count(fips, year, name = "n_plans")

ggplot(county_plan_counts, aes(x = factor(year), y = n_plans)) +
  geom_boxplot() +
  labs(
    x = "Year",
    y = "Number of MA Plans",
    title = "Distribution of MA Plan Counts by County Over Time"
  ) +
  theme_minimal()
```

```r
# Plan bid histogram
plan.bids <- data %>%
  left_join(risk_rebate, by = c("planid", "year")) %>%
  mutate(
    bid = premium + rebate
  )

plan.bids %>%
  filter(year %in% c(2014, 2018)) %>%
  ggplot(aes(x = bid)) +
  geom_histogram(bins = 40, fill = "steelblue", alpha = 0.7) +
  facet_wrap(~ year, scales = "free_x") +
  labs(
    x = "Plan Bid",
    y = "Count",
    title = "Distribution of MA Plan Bids"
  ) +
  theme_minimal()
```

```r
# HHI over time
county_hhi <- data %>%
  left_join(ma.penetration, by = c("fips", "year")) %>%
  group_by(fips, year) %>%
  mutate(
    share = enrollment / sum(enrollment)
  ) %>%
  summarize(
    hhi = sum(share^2),
    .groups = "drop"
  )
```

```r
county_hhi %>%
  group_by(year) %>%
  summarize(avg_hhi = mean(hhi, na.rm = TRUE)) %>%
  ggplot(aes(x = year, y = avg_hhi)) +
  geom_line() +
  geom_point() +
  labs(
    x = "Year",
    y = "Average HHI",
    title = "Average MA Market Concentration Over Time"
  ) +
  theme_minimal()
```

```r
# MA penetration over time
ma.penetration %>%
  group_by(year) %>%
  summarize(
    ma_share = sum(ma_enrollment) / sum(total_medicare)
  ) %>%
  ggplot(aes(x = year, y = ma_share)) +
  geom_line() +
  geom_point() +
  labs(
    x = "Year",
    y = "MA Share",
    title = "Medicare Advantage Penetration Over Time"
  ) +
  theme_minimal()
```
